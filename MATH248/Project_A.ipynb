## Naive Approaches to check if number is prime
The Naive solution to checking if a number is prime divides the number n by every number smaller than n but greater than 1. If at anytime the divison results in an integer then number has a divisior other than 1 and n so n is not prime and the function immediately returns false. If the divison never results in an integer then n is prime and thus the function returns true. Any value equal to or less than 1 is not prime so the function immedietly returns false.

```python
# Naive solution
def isPrime(n):
    if n <= 1:
        return False
    divisor = n - 1
    while divisor > 1:
        m = n / divisor
        if m.is_integer():
            return False
        divisor -= 1
    return True
```

```python
for i in range(0,101):
    print(f"  {i}: {isPrime(i)}", end = "")
```

### Improvements
Improvements can be made by using the modulus operator instead of basic division because the result of the division for this function does not matter.The modulus just returns the remainder which is the only value that needs to be checked to see if a number is divisiable by number other than 1 and n. All vaues less than n do not need to be checked only values less than $\sqrt{n}$. The runtime is O(n).

## Naive Approach version 2 Iteration
This solution performs logically identically to the one above but instead of using divison it uses modulus to check if the remainder is 0 and  if it is at anypoint then the function will return false. If the remainder is never 0 then the function returns true. Also only values less than $\sqrt{n}$ are checked as divisors.

```python
# Iterative solution
def isPrimeIteration(n):
    if n <= 1:
        return False
    divisor = int(n ** 1/2)
    while divisor > 1:
        if n % divisor == 0:
            return False
        divisor -= 1
    return True
```

```python
for i in range(0,101):
    print(f"  {i}: {isPrimeIteration(i)}", end = "")
```

### Improvements
There are no obvious improvements to the code but a second implementation using recusion in place of iteration can be implemented. The runtime is O($\sqrt{n}$).

## Naive Approach Version 2.1
This solution is the recursive solution to the checking if a number is prime. Logically it performs indentical to the naive approach version 2 but with the main difference being the use of recursion instead of iteration.

```python
## Recursive solution
def isPrimeRecursion(n):
    if n <= 1:
        return False
    def primeRecursion(divisor):
        if divisor <= 1:
            return True
        if n % divisor == 0:
            return False
        return primeRecursion(divisor - 1)
    return primeRecursion(int(n ** 1/2))
```

```python
for i in range(0,101):
    print(f"  {i}: {isPrimeRecursion(i)}", end = "")
```

### Improvements
There are no obvious improvments maybe some changes in if statement locations but nothing that would change the runtime drastically. The runtime is O($\sqrt{n}$).

## Naive Approach to Sieve of Eratosthenes
This approach creates a dictonary of all values from 2 to n. The loop checks if a value is a multiple of a prime smaller than or equal to it and if it is then that number is not prime so its dictonary value is set to false. After completionof the loop all values in the dictionary that are not false are returned.

```python
# Sieve of Eratosthenes Naive approach
def soe(n):
    nums = dict()
    for i in range(2, n + 1):
        nums.update({i:True})
    p = 2
    while p <= n:
        if nums.get(p):
            i = 2
            while p * i <= n:
                nums.update({p * i:False})
                i += 1
        p += 1
    result = []
    for x in nums:
        if nums.get(x):
            result.append(x)
    return result
```

```python
arr = soe(121)
print(arr)
```

### Improvements
This approach can be improved by limiting the number of values that p checks by checking values less than $\sqrt{n}$. The time complexity is O($n^2$).

## Improved Approach to Sieve of Eratosthenes
This approach follows the same logic as above but with the improvement of checking numbers less than $\sqrt{n}$.

```python
# Improved Version
def sieve(n):
    nums = dict()
    for i in range(2, n + 1):
        nums.update({i:True})
    for i in range(2, int(math.sqrt(n)) + 1):
        if nums.get(i):
            i2 = i ** 2
            j = 0
            while i2 + j*i <= n:
                nums.update({i2 + j*i: False})
                j += 1
    result = []
    for x in nums:
        if nums.get(x):
            result.append(x)
    return result
```

```python
arr = sieve(121)
print(arr)
```

### Improvements
There are no obvious improvements to this approach. The time complexity is $O(n log log n)$.

## Naive Approach to Conway’s Prime Producing Machine
This approach uses an array of floats to hold the fractions for conways prime game. This appraoch while it does give the first 3 primes it cannot achive any higher primes due to faults in the machine epsilon cause small imprefections in the fraction calculations eventually resulting in loss of precision. This loss of precision causes future calcualtion to be off.

```python
def conway(steps):
    primes = []
    fractions = np.array([17/91, 78/85, 19/51, 23/38, 29/33, 77/29, 95/23, 77/19, 1/17, 11/13, 13/11, 15/14, 15/2, 55], dtype = np.double)
    n = np.double(2)
    i = 1
    while i <= steps:
        for fraction in fractions:
            m = np.double(n * fraction)
            if math.isclose(m, int(m), rel_tol=1e-12):
                n = int(m)
                break
            elif math.isclose(m, (int(m) + 1), rel_tol=1e-12):
                n = int(m) + 1
                break

        p = n
        k = 0
        while p > 1:
            p /= 2
            k += 1
        if p == 1:
            primes.append(k)
        i += 1
    return primes
```

```python
print(conway(8102))
```

### Improvements
An improvment could be to use two arrays to store the numerators and denominators for the fraction thus elimination floats and using only integers.

## Improved Approach to Conway's Prime Producing Machine
This approach uses two arrays to store the numerators and denominators this allows calcualtions to be done with int values insted of float values. This approach while calcualting more primes still faults after the prime number 7 is found. Further exploration of this problem is needed to find out why.

```python
def conwayv2(steps):
    primes = []
    numerators = np.array([17, 78, 19, 23, 29, 77, 95, 77, 1, 11, 13, 15, 15, 55])
    denominators = np.array([91, 85, 51, 38, 33, 29, 23, 19, 17, 13, 11, 14, 2, 1])
    n = 2
    i = 1
    while i <= steps:
        for j in range(14):
            m = (n * numerators[j]) / denominators[j]
            if m.is_integer():
                n = int(m)
                break
        #print(f"{i}: {n}")
        p = n
        k = 0
        while p > 1:
            p /= 2
            k += 1
        if p == 1:
            primes.append(k)
        i += 1
    return primes
```

```python
print(conwayv2(1000))
```

### Improvements
There are no obvious improvements.

Conway's Prime counting machine can run infinitly given an infite number of steps and should produce an infite number of primes. The first approach will not produce an infite number due to there being error in float arithmetic causing it not to produce all primes. The second approach should produce all primes as it eliminates the need for float storage and aritmetic effectivly limiting error due to float arithmetic. Both approahces run in O(n) time where n is the number of steps.
